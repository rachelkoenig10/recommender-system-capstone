```python
# Import Tokenizer
from nltk.tokenize import RegexpTokenizer

tokenizer = RegexpTokenizer(r'\w+')
```

```python
users['asins'] = users['asins'].apply(lambda x: tokenizer.tokenize(x))
```

```python
list_of_asins = {asin for user in users['asins'] for asin in user}
```

```python
type(list_of_asins)
```

```python
import time 

# Begin timer.
t0 = time.time()
# Instantiate counter (for progress statement).
count = 0
# Iterate through list of colors.
for asin in list_of_asins:
    # Print out progress statement.
    if count % 100 == 0:
        print("Completed " + str(count) + " out of " + str(len(list_of_asins)) + \
            " columns in " + str(round(time.time() - t0,4)) + " seconds.")
        
        
# Create column for each color.
# Each value is 1 if the product is represented by that color .
# Each value is 0 if the product does not have that color in the description.        
    users[asin] = [1 if asin in users.loc[row, 'asins'] else 0 for row in users.index]
# Add one to counter (for progress statement).
    count += 1

# Code adapted from Matt Brems
```

```python
X = users.drop(columns= ['reviewerID', 'names', 'asins', 'colors'])
```

```python
ss = StandardScaler()
X_scaled = ss.fit_transform(X)
```

```python
km = KMeans(n_clusters=3, random_state=42)
km.fit(X_scaled)
```

```python
km.cluster_centers_
```

```python
km.labels_
```

```python
km.predict()
```

```python
users['clusters'] = km.labels_
```

```python
users['clusters'].value_counts()
```

```python
dbscan = DBSCAN(eps=2.3, min_samples=4)
dbscan.fit(X_scaled)
```

```python
dbscan.labels_
```

```python
silhouette_score(X_scaled, dbscan.labels_)
```

```python
users['DBSCAN'] = dbscan.labels_
users.head()
```

```python
users['DBSCAN'].value_counts()
```
